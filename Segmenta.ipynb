Checking the description and IDs match up

```python
## Get a set of the unique names

uniques = set(df['Description'])

## Find out how long it is

len(uniques)
```

Striping the description column as they dont match 

```python
## Strip spaces and lower the casing into stripped uniques set

## Have our original 'uniques' set
uniques = set(df['Description'])

## Make an empty set
strip_unique = set()

## Loop through the items in the original set and add their lower-case and stripped versions to our empty set
for item in uniques:
    strip_unique.add(item.strip().lower())

## Let's check the length of the new set! 
len(strip_unique)
```
